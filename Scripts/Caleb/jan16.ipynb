### Arnulf DNA from folder 4


```python
folderPath1 = r"C:\Users\Perkins Lab\Desktop\Caleb Maddry\Data\2025\January\16\ArnulfDNA\4\DataTSV"

dataFrame = openFiles(folderPath1, numRef=2, numExp=4, fps=400)
```

```python
ref = dataFrame["zRefBead1"]
exp = dataFrame["zExpBead1"]
time = dataFrame["timeVector"]

glassPosition = -270

window_width = 100

cumsum_vec = np.cumsum(np.insert(exp, 0, 0)) 
smoothExp = (cumsum_vec[window_width:] - cumsum_vec[:-window_width]) / window_width

cumsum_vec = np.cumsum(np.insert(ref, 0, 0)) 
smoothRef = (cumsum_vec[window_width:] - cumsum_vec[:-window_width]) / window_width

plt.plot(exp - ref - glassPosition, color="#adebad")
plt.plot(smoothExp - smoothRef - glassPosition, color="#145214")
plt.ylim(-50,1000)
plt.xlim(0, 40000)
plt.title("Z data for Bead 1"); plt.xlabel("Frame # (400 fps)"); plt.ylabel("Extension (nm)")
plt.show()

plt.hist(exp[0:13000] - ref[0:14000], bins=150, color="#145214");

plt.axvline(glassPosition, color="red")
plt.show()
```
