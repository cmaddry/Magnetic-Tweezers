```python
import matplotlib.pyplot as plt
import numpy as np
from calibrationFunctions import openFiles, calculateForce, calculateTetherExtension, findTetheredBeads
```

```python
folderPath = r"C:\Users\Perkins Lab\Desktop\Caleb Maddry\MT Data Analysis\Magnetic-Tweezers\Data\December\19\ArnulfPS"

dataFrame = openFiles(folderPath, numRef=0, numExp=10, fps=400);
```

```python
window_width = 100

bead = dataFrame["zExpBead1"]

cumsum_vec = np.cumsum(np.insert(bead, 0, 0)) 
smoothExp = (cumsum_vec[window_width:] - cumsum_vec[:-window_width]) / window_width

plt.plot(bead[0:4000] - np.mean(bead[0:4000]))
plt.plot(smoothExp[0:4000] - np.mean(smoothExp[0:4000]))
plt.title("Bead 1"); plt.xlabel("Frame # (400 fps)"); plt.ylabel("nm");
plt.show()

plt.hist(bead[0:4000], bins=100);
plt.title("Bead 1"); plt.xlabel("nm"); plt.ylabel("Count");
plt.show()
```
