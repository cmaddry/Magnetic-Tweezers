```python
import matplotlib.pyplot as plt
import numpy as np

# Choice for all of the start and stop times of the experiment
highForceStart = 0;
highForceEnd = 24000;

medForceStart = 27200
medForceEnd = 40000

lowForceStart = 50000
lowForceEnd = 90000

forceExtStart =110000
forceExtEnd = len(dataFrame["linearMotor"])

# Creation of the vectors needed for calculations and plotting
time = dataFrame["timeVector"]
extension = dataFrame["zExpBead2"] - 5730
xData = dataFrame["xExpBead2"] 
yData = dataFrame["yExpBead2"] 


## Plotting
# Extension vs time
plt.plot(time, extension,
         color="magenta")
plt.axhline(0,
            color="red",
            label="x = 0")
plt.title("Extension for full experiment")
plt.ylabel("Extension (nm)")
plt.xlabel("Time (s)")
plt.show()

# Histogram of the extension at low force
plt.hist(extension[lowForceStart:lowForceEnd], 
         bins=100,
         color="magenta");
plt.ylabel("Count")
plt.xlabel("Extension (nm)")
plt.title("Histogram at low force")
plt.show()
```

```python
print("Variance of y-data:", np.var(yData[lowForceStart:lowForceEnd]), 
      "\nVariance of x-data:", np.var(xData[lowForceStart:lowForceEnd])) # X data is always gonna be lower


a = highForceStart
b = highForceEnd

force = 4.1*np.mean(extension[a:b])/np.var(xData[a:b])

print("\nThe calculated force is:", force,"pN")


fig, ax = plt.subplots(1, 2, figsize = (20,5));
ax[0].plot(time[a:b], extension[a:b],
           color='#5ADD4D',
           label="Extension");
ax[1].scatter(xData[a:b] - np.mean(xData[a:b]), yData[a:b] - np.mean(yData[a:b]),
              color='#1BcD9A');
plt.title("X vs. Y Data")
plt.xlim(-550,650);
plt.ylim(-550,650);
```
